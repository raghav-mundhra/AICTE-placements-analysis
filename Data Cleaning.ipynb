```python
import pandas as pd
import numpy as np
from impyute.imputation.cs import mice
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df.dtypes
```

## Applying MICE for missing value handeling

```python
# Filling missing values using MICE algorithm
train_temp_mice_Impyute = df.drop(columns=['Year'])
train_cols = list(train_temp_mice_Impyute)
imputed_training_Impyute = pd.DataFrame(mice(train_temp_mice_Impyute.values))
imputed_training_Impyute.columns = train_cols
```

```python
df[train_cols] = imputed_training_Impyute[train_cols]
df['Students Passed'] = df['Students Passed'].astype(int)
df
```

## Creating a column: Placement Ratio

Placement Ratio = Placement/Students Passed

```python
df['Placement Ratio'] = df['Placement']/df['Students Passed']
```
